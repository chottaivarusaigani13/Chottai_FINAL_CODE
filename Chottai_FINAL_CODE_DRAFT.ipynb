```python
# Indicator Selection
import pandas as pd

selected_indicators = [
    'NY.GDP.MKTP.KD.ZG',   # GDP growth
    'SP.POP.GROW',         # Population growth
    'FP.CPI.TOTL.ZG',      # Inflation
    'SP.DYN.LE00.IN',      # Life expectancy
    'EN.ATM.CO2E.PC'       # CO2 emissions
]

# Filter dataset
df_filtered = df[df['Indicator Code'].isin(selected_indicators)]

# Check result
print("Filtered Dataset Shape:", df_filtered.shape)

df_filtered.head()
```

```python
# Identify Year Columns

year_columns = [col for col in df_filtered.columns if col.isdigit()]

print("Number of Year Columns:", len(year_columns))
print("Sample Years:", year_columns[:10])
```

```python
# Convert to Long Format

df_long = df_filtered.melt(
    id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],
    value_vars=year_columns,
    var_name='Year',
    value_name='Value'
)

df_long.head()
```

```python
# Convert Year Column to Integer
df_long['Year'] = df_long['Year'].astype(int)
```

```python
# Country-wise Sorting
df_long = df_long.sort_values(
    by=['Country Code', 'Indicator Code', 'Year']
)

df_long.head()
```

```python
# Reset Index
df_long = df_long.reset_index(drop=True)

df_long.head()
```

```python
# Check Dataset Structure
df_long.info()
```

```python
# Missing Values Before

print("Missing Values Before:", df_long['Value'].isnull().sum())
```

```python
# Interpolation

df_long['Value'] = df_long.groupby(
    ['Country Code', 'Indicator Code']
)['Value'].transform(lambda x: x.interpolate(method='linear'))
```

```python
# Forward Fill

df_long['Value'] = df_long.groupby(
    ['Country Code', 'Indicator Code']
)['Value'].transform(lambda x: x.fillna(method='ffill'))
```

```python
# Backward Fill

df_long['Value'] = df_long.groupby(
    ['Country Code', 'Indicator Code']
)['Value'].transform(lambda x: x.fillna(method='bfill'))
```

```python
# Verify Missing Values After Treatment
print("Missing Values After:", df_long['Value'].isnull().sum())
```

```python
# Drop Remaining NaN Values

df_long = df_long.dropna(subset=['Value'])

print("Remaining Missing Values:", df_long['Value'].isnull().sum())
print("Dataset Shape After Cleaning:", df_long.shape)
```

```python
df_long.head()
```

```python
# Feature Engineering

# Lag features (previous years)
df_long['lag_1'] = df_long.groupby(
    ['Country Code', 'Indicator Code']
)['Value'].shift(1)

df_long['lag_2'] = df_long.groupby(
    ['Country Code', 'Indicator Code']
)['Value'].shift(2)

# Growth rate
df_long['growth_rate'] = df_long.groupby(
    ['Country Code', 'Indicator Code']
)['Value'].pct_change()

# Rolling mean (3-year)
df_long['rolling_mean_3'] = df_long.groupby(
    ['Country Code', 'Indicator Code']
)['Value'].transform(lambda x: x.rolling(window=3).mean())

df_long.head()
```

```python
# Remove Rows Created by Lag Operations

df_long = df_long.dropna()

print("Dataset Shape After Feature Engineering:", df_long.shape)
```

```python
# Fix Infinite Values Before Scaling

# Replace infinity values with NaN
df_long.replace([np.inf, -np.inf], np.nan, inplace=True)

# Drop rows containing NaN
df_long = df_long.dropna()

print("Dataset Shape After Removing Infinite Values:", df_long.shape)
```

```python
# Define Features and Target

feature_columns = [
    'lag_1',
    'lag_2',
    'growth_rate',
    'rolling_mean_3'
]

X = df_long[feature_columns]
y = df_long['Value']

print("Feature Matrix Shape:", X.shape)
print("Target Variable Shape:", y.shape)
```

```python
# Time-Based Split

train_data = df_long[df_long['Year'] <= 2015]
test_data = df_long[df_long['Year'] > 2015]

X_train = train_data[feature_columns]
y_train = train_data['Value']

X_test = test_data[feature_columns]
y_test = test_data['Value']

print("Training Set:", X_train.shape)
print("Testing Set:", X_test.shape)
```

```python
# Verify Year Split
print("Training Years:", train_data['Year'].min(), "-", train_data['Year'].max())
print("Testing Years:", test_data['Year'].min(), "-", test_data['Year'].max())
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled = pd.DataFrame(X_train_scaled, columns=feature_columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=feature_columns)

X_train_scaled.head()
print("Scaled Training Shape:", X_train_scaled.shape)
print("Scaled Testing Shape:", X_test_scaled.shape)
```

```python
# Model Training

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

# Initialize models
lr_model = LinearRegression()
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
gb_model = GradientBoostingRegressor(n_estimators=100, random_state=42)

# Train models
lr_model.fit(X_train_scaled, y_train)
rf_model.fit(X_train_scaled, y_train)
gb_model.fit(X_train_scaled, y_train)

print("Models trained successfully!")
```

```python
# Model Predictions

lr_pred = lr_model.predict(X_test_scaled)
rf_pred = rf_model.predict(X_test_scaled)
gb_pred = gb_model.predict(X_test_scaled)
```

```python
# Model Evaluation

from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# Linear Regression
lr_rmse = np.sqrt(mean_squared_error(y_test, lr_pred))
lr_mae = mean_absolute_error(y_test, lr_pred)

# Random Forest
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_pred))
rf_mae = mean_absolute_error(y_test, rf_pred)

# Gradient Boosting
gb_rmse = np.sqrt(mean_squared_error(y_test, gb_pred))
gb_mae = mean_absolute_error(y_test, gb_pred)

print("Linear Regression RMSE:", lr_rmse)
print("Linear Regression MAE:", lr_mae)

print("Random Forest RMSE:", rf_rmse)
print("Random Forest MAE:", rf_mae)

print("Gradient Boosting RMSE:", gb_rmse)
print("Gradient Boosting MAE:", gb_mae)
```

```python
# Model Comparison Table

results = pd.DataFrame({
    'Model': ['Linear Regression', 'Random Forest', 'Gradient Boosting'],
    'RMSE': [lr_rmse, rf_rmse, gb_rmse],
    'MAE': [lr_mae, rf_mae, gb_mae]
})

results
```

```python
# Select Time Series

ts_data = df_long[
    (df_long['Country Code'] == 'IND') &
    (df_long['Indicator Code'] == 'NY.GDP.MKTP.KD.ZG')
][['Year','Value']]

ts_data = ts_data.sort_values('Year')

# Fix index (IMPORTANT)
ts_data = ts_data.set_index('Year')
ts_data.index = pd.to_datetime(ts_data.index, format='%Y')
ts_data.head()
```

```python
# ARIMA Model

from statsmodels.tsa.arima.model import ARIMA

# Fit ARIMA model
arima_model = ARIMA(ts_data['Value'], order=(2,1,2))
arima_result = arima_model.fit()

# Forecast next 10 years
arima_forecast = arima_result.forecast(steps=10)

print(arima_forecast)
```

```python
# ARIMA Model Evaluation

from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# In-sample predictions
arima_pred = arima_result.predict(start=1, end=len(ts_data)-1)

# Align actual values
actual_values = ts_data['Value'][1:]

# Calculate metrics
arima_rmse = np.sqrt(mean_squared_error(actual_values, arima_pred))
arima_mae = mean_absolute_error(actual_values, arima_pred)

print("ARIMA RMSE:", arima_rmse)
print("ARIMA MAE:", arima_mae)
```

```python
# ARIMA Actual vs Predicted Plot

plt.figure(figsize=(12,6))

plt.plot(ts_data.index[1:], actual_values, label='Actual')
plt.plot(ts_data.index[1:], arima_pred, label='ARIMA Predicted')

plt.title("ARIMA Model: Actual vs Predicted")
plt.xlabel("Year")
plt.ylabel("Indicator Value")

plt.legend()
```

```python
# Future Forecast Visualization

future_years = np.arange(ts_data.index.max().year + 1,
                         ts_data.index.max().year + 11)

plt.figure(figsize=(12,6))

# Historical data
plt.plot(ts_data.index.year, ts_data['Value'], label='Historical Data')

# Forecast
plt.plot(future_years, arima_forecast, label='ARIMA Forecast', linestyle='dashed')

plt.title("ARIMA Future Forecast")
plt.xlabel("Year")
plt.ylabel("Indicator Value")

plt.legend()
plt.show()
```

```python
# Combined Historical + Forecast Plot (Best for Dissertation)
plt.figure(figsize=(12,6))

plt.plot(ts_data.index.year, ts_data['Value'], label='Actual')

combined_years = list(ts_data.index.year) + list(future_years)
combined_values = list(ts_data['Value']) + list(arima_forecast)

plt.plot(combined_years, combined_values, label='Forecast Trend', linestyle='dashed')

plt.title("Historical and Forecasted Indicator Trend (ARIMA)")
plt.xlabel("Year")
plt.ylabel("Indicator Value")

plt.legend()
plt.show()
```

```python
# LSTM Model

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
```

```python
# Scale Data
scaler = MinMaxScaler()

ts_scaled = scaler.fit_transform(ts_data[['Value']])
```

```python
# Create Sequences
def create_sequences(data, window=3):

    X = []
    y = []

    for i in range(len(data)-window):
        X.append(data[i:i+window])
        y.append(data[i+window])

    return np.array(X), np.array(y)

X_lstm, y_lstm = create_sequences(ts_scaled)
```

```python
# Train LSTM Model
model_lstm = Sequential()

model_lstm.add(LSTM(50, activation='relu', input_shape=(X_lstm.shape[1],1)))
model_lstm.add(Dense(1))

model_lstm.compile(optimizer='adam', loss='mse')

model_lstm.summary()
```

```python
model_lstm.fit(X_lstm, y_lstm, epochs=50, verbose=1)
```

```python
# LSTM Prediction
lstm_pred = model_lstm.predict(X_lstm)

lstm_pred = scaler.inverse_transform(lstm_pred)
```

```python
# LSTM Model Evaluation

# Actual values corresponding to predictions
actual_lstm = scaler.inverse_transform(y_lstm)

# Calculate metrics
lstm_rmse = np.sqrt(mean_squared_error(actual_lstm, lstm_pred))
lstm_mae = mean_absolute_error(actual_lstm, lstm_pred)

print("LSTM RMSE:", lstm_rmse)
print("LSTM MAE:", lstm_mae)
```

```python
# LSTM Actual vs Predicted Visualization
plt.figure(figsize=(12,6))

plt.plot(actual_lstm, label='Actual')
plt.plot(lstm_pred, label='LSTM Predicted')

plt.title("LSTM Model: Actual vs Predicted")
plt.xlabel("Time Step")
plt.ylabel("Indicator Value")

plt.legend()
plt.show()
```

```python
# Time-Series Visualization with Years
plt.figure(figsize=(12,6))

years = ts_data.index.year[3:]   # offset because of sequence window

plt.plot(years, actual_lstm, label='Actual')
plt.plot(years, lstm_pred, label='LSTM Prediction')

plt.title("LSTM Forecasting Performance")
plt.xlabel("Year")
plt.ylabel("Indicator Value")

plt.legend()
plt.show()
```

```python
# LSTM Future Forecast

future_steps = 10
last_sequence = ts_scaled[-3:]

future_predictions = []

for i in range(future_steps):

    pred = model_lstm.predict(last_sequence.reshape(1,3,1))
    future_predictions.append(pred[0][0])

    last_sequence = np.append(last_sequence[1:], pred)

future_predictions = scaler.inverse_transform(np.array(future_predictions).reshape(-1,1))
```

```python
# Future Forecast Visualization
future_years = np.arange(
    ts_data.index.year.max() + 1,
    ts_data.index.year.max() + 11
)

plt.figure(figsize=(12,6))

plt.plot(ts_data.index.year, ts_data['Value'], label='Historical Data')
plt.plot(future_years, future_predictions, label='LSTM Forecast', linestyle='dashed')

plt.title("LSTM Future Forecast")
plt.xlabel("Year")
plt.ylabel("Indicator Value")

plt.legend()
plt.show()
```
